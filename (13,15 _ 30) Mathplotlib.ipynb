```python
import matplotlib.pyplot as plt  
import numpy as np
from sklearn.datasets import load_iris
```

# Cours 1 

```python
x = np.linspace(0,2,10)
y = x**2
```

```python
plt.figure(figsize=(7,5))
plt.plot(x,y, label='qudratic')
plt.plot(x,x**3, label='cubic')
plt.title('Fig1')
plt.xlabel('AxeX')
plt.ylabel('AxeY')
plt.legend()
plt.show()
plt.savefig('figure.jpeg')
```

```python
plt.figure(figsize=(8,5))
plt.subplot(2,2,1)
plt.plot(x,x**2, c='black', label='Quadratic')
plt.legend()
plt.title('Grapique1')
plt.subplot(2,2,2)
plt.plot(x,x**3, c='yellow', label='Cubic')
plt.legend()
plt.title('Graphic2')
plt.show()
```

### Coutour plot ~ Solution d'optimisation

```python
f = lambda x, y : np.sin(x) + np.cos(x+y)*np.cos(x)
```

```python
X = np.linspace (0, 5, 100)
Y = np.linspace (0, 5, 100)
X, Y = np.meshgrid(X, Y) 
# np.meshrid permet de créer un grille à partir de 2 axes

Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, 20)
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, 20, cmap='cubehelix')
plt.colorbar()
```
