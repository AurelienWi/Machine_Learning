### Resample

```python
# Permet de regrouper nos données selon une fréquence temporelle
## Qui dit groupe de données dit statistique, par mois, par jour, etc... avec mean(), std(), max(), min(), etc..

### On obtient ici la moyenne par mois du prix du bitcoin en 2019

df.loc['2019','Close'].resample('M').mean().plot()
```

```python
plt.figure(figsize=(12,9))
df.loc['2019','Close'].plot()
df.loc['2019','Close'].resample('W').mean().plot(label='Moyenne par semaine', lw=2, c='green', ls='--', alpha=0.8)
df.loc['2019','Close'].resample('M').mean().plot(label='Moyenne par mois', lw=3, c='orange', ls=':')

plt.legend()
```

```python
df.loc['2019','Close'].resample('M').agg(['mean', 'std', 'min', 'max']).plot()
```

### Moving Average

```python
df.loc['2019','Close'].rolling(window=7).mean(). plot()
```

```python
plt.figure(figsize=(12,9))
df.loc['2019-09','Close'].plot()
df.loc['2019-09','Close'].rolling(window=7).mean(). plot(label='MA(7)', lw=3, ls='dashdot', alpha=0.7)
df.loc['2019-09','Close'].rolling(window=7, center=True).mean(). plot(label='MA(7)', lw=3, ls='dashdot', alpha=0.7)

plt.legend()
```

### Moyenne Mobile Exponentielle _ EWM

```python
df.loc['2019-09','Close'].plot()

df.loc['2019-09','Close'].ewm(alpha=0.5).mean(). plot(label='MA(7)', lw=3, ls='dashdot', alpha=0.7)
```
